### **Modelos de ML Supervisados:** <a class="tocSkip">
Los modelos supervisados se entrenan utilizando un conjunto de datos etiquetado, donde cada ejemplo de entrada está asociado con su correspondiente etiqueta o resultado deseado. El objetivo del modelo es aprender a mapear las entradas a las salidas correctas utilizando patrones identificados en los datos de entrenamiento [[2]](#r2). Luego, el modelo puede realizar predicciones o clasificaciones precisas en nuevas instancias no vistas. 

### **Modelos de ML No Supervisados:** <a class="tocSkip">
Los modelos no supervisados se utilizan cuando no hay etiquetas o resultados conocidos en el conjunto de datos de entrenamiento. En este enfoque, el modelo busca descubrir patrones o estructuras ocultas en los datos sin tener una guía explícita sobre lo que se busca. Los modelos no supervisados a menudo se utilizan para tareas de agrupamiento (clusterización) y reducción de dimensionalidad [[2]](#r2), donde el objetivo es organizar datos similares en grupos o simplificar la representación de datos complejos. 

### Modelos por Regresión:<a class="tocSkip">
La regresión es un tipo de análisis estadístico que se utiliza para predecir valores numéricos (continuos) en función de variables independientes. Los modelos de regresión buscan establecer una relación matemática entre las variables de entrada y la variable objetivo, de modo que se pueda hacer una estimación o predicción de valores futuros [[2]](#r2). Estos modelos son útiles cuando se busca comprender cómo una variable dependiente cambia en función de una o más variables independientes.

### Modelos por Clasificación: <a class="tocSkip">
Los modelos de clasificación son utilizados para asignar etiquetas o categorías a datos en función de sus características. En lugar de predecir un valor numérico como en la regresión, los modelos de clasificación asignan una clase a una observación basándose en patrones identificados en los datos de entrenamiento [[2]](#r2). Por ejemplo, se podría usar un modelo de clasificación para predecir si un correo electrónico es spam o no spam, o para clasificar imágenes de animales en diferentes categorías.

### Clustering: <a class="tocSkip">
El clustering es un enfoque de aprendizaje no supervisado que se utiliza para agrupar datos en grupos o clústeres. A diferencia de la regresión y la clasificación, el clustering no se basa en etiquetas previas ni en una variable objetivo. En cambio, busca identificar patrones y similitudes en los datos y agruparlos en clústeres basados en su similitud [[2]](#r2). Estos clústeres pueden ayudar a descubrir estructuras ocultas en los datos y proporcionar una forma de comprender la organización intrínseca de los mismos.

###  Regresión lineal

La regresión lineal es una técnica de aprendizaje supervisado que se utiliza para modelar la relación entre una variable dependiente y una o más variables independientes. Su objetivo es encontrar la mejor línea recta (en el caso unidimensional) o un hiperplano (en el caso multidimensional) que represente la relación lineal entre las variables de manera óptima.[[3]](#r3)

**Modelo de Regresión Lineal Simple:**

El caso más simple es la regresión lineal simple, donde se tiene una sola variable independiente y una variable dependiente. El modelo puede expresarse como:

y = β₀ + β₁ ⋅ x + ε


Donde:
- \(y\) es la variable dependiente (o variable objetivo).
- \(x\) es la variable independiente (o variable de entrada).
- \(β₀\) y \(β₁\) son los coeficientes del modelo, que representan la intersección y la pendiente de la línea recta, respectivamente.
- \(ε\) es el término de error, que representa la diferencia entre los valores reales y los predichos por el modelo.

El objetivo en la regresión lineal es encontrar los valores óptimos de \(β₀\) y \(β₁\) para minimizar el error cuadrático medio entre las predicciones y los valores reales[[3]](#r3).

**Modelo de Regresión Lineal Múltiple:**

En el caso de la regresión lineal múltiple[[3]](#r3), se tienen varias variables independientes 
($x_1$, $x_2$, $\ldots$, $x_n$)
y una variable dependiente \(y\). El modelo se generaliza como:

\[ y = $\beta_0 + \beta_1 \cdot x_1 + \beta_2 \cdot x_2 + \ldots + \beta_n \cdot x_n + \varepsilon$ \]

El objetivo sigue siendo encontrar los valores óptimos de los coeficientes (\($\beta_0$, $\beta_1$, $\ldots$, $\beta_n$\)) que minimicen el error entre las predicciones y los valores reales.


**Método de los Mínimos Cuadrados:**

El método más común para encontrar los coeficientes óptimos en la regresión lineal es el método de los mínimos cuadrados. Este método minimiza la suma de los cuadrados de las diferencias entre las observaciones reales y las predicciones del modelo.[[3]](#r3)

**Evaluación de la Regresión Lineal:**

Una vez que se ha ajustado el modelo, es importante evaluar su rendimiento. Algunas métricas comunes para evaluar la regresión lineal son el coeficiente de determinación (R-cuadrado), el error cuadrático medio (MSE) y el error absoluto medio (MAE).

**Aplicaciones de la Regresión Lineal:**

La regresión lineal es ampliamente utilizada en diversos campos, incluyendo ciencias sociales, economía, finanzas, ingeniería y ciencias naturales. Algunas de sus aplicaciones incluyen:
- Predicción de precios de viviendas basado en características como el área, ubicación, etc.
- Análisis de tendencias en datos económicos.
- Estimación de la demanda de productos.
- Modelado de relaciones entre variables físicas en ciencias e ingeniería.



**Regresión Lineal Simple: Predicción en conversión de grados Celsius a Fahrenheit**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import time

# Datos de temperatura en Celsius y Fahrenheit
celsius = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
fahrenheit = np.array([32, 41, 50, 59, 68, 77, 86, 95, 104, 113, 122, 131, 140, 149, 158])
# print(celsius)
# Cambiar la forma de los datos para que sklearn los pueda usar
celsius = celsius.reshape(-1, 1)
# print(celsius)
fahrenheit = fahrenheit.reshape(-1, 1)

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Obtener el tiempo de inicio del entrenamiento
inicio_entrenamiento = time.time()

# Entrenar el modelo con los datos de entrenamiento
modelo.fit(celsius, fahrenheit)

# Obtener el tiempo de finalización del entrenamiento
fin_entrenamiento = time.time()

# Obtener los coeficientes del modelo
# Obtiene la intercepción con el eje y 
beta_0 = modelo.intercept_[0]
# Obtiene la intercepción con el eje x
beta_1 = modelo.coef_[0][0]

# Calcular las predicciones del modelo para los datos de entrenamiento
predicciones = modelo.predict(celsius)

# Calcular el error cuadrático medio
error_cuadratico_medio = mean_squared_error(fahrenheit, predicciones)

# Calcular el coeficiente de determinación R2
r2 = r2_score(fahrenheit, predicciones)

# Imprimir los resultados
print("Coeficiente beta_0 (intercepto):", beta_0)
print("Coeficiente beta_1 (pendiente):", beta_1)
print("Error cuadrático medio:", error_cuadratico_medio)
print("Coeficiente de determinación R2:", r2)
print("Duración del entrenamiento (segundos):", fin_entrenamiento - inicio_entrenamiento)

# Realizar una predicción para una temperatura en Celsius específica
temperatura_celsius = 100
temperatura_fahrenheit_prediccion = modelo.predict([[temperatura_celsius]])
print(f"La temperatura {temperatura_celsius} °C es equivalente a {temperatura_fahrenheit_prediccion[0][0]:.2f} °F.")

# Crear la gráfica
plt.scatter(celsius, fahrenheit, label='Datos', color='blue')
plt.plot(celsius, predicciones, label='Tendencia lineal', color='red')
plt.scatter(temperatura_celsius, temperatura_fahrenheit_prediccion, color='green', label='Predicción')
plt.xlabel('Temperatura Celsius (°C)')
plt.ylabel('Temperatura Fahrenheit (°F)')
plt.title('Regresión Lineal: Temperaturas Celsius a Fahrenheit')
plt.legend()
plt.grid(True)
plt.show()
````

###  Regresión Polinomial

La regresión polinomial es una técnica de regresión que extiende el modelo de regresión lineal al utilizar una relación polinomial entre la variable independiente (X) y la variable dependiente (Y). En lugar de ajustar una línea recta a los datos, la regresión polinomial ajusta una curva polinomial para capturar mejor las relaciones no lineales en los datos[[4]](#r4).

En un modelo de regresión polinomial, la relación entre X e Y se representa mediante una función polinomial de grado n, donde n es un número entero positivo que determina el grado del polinomio[[4]](#r4). La ecuación general de un modelo de regresión polinomial de grado n es:

```
Y = β₀ + β₁*X + β₂*X² + ... + βₙ*Xⁿ + ε
```

Donde:
- **Y** es la variable dependiente (la variable que queremos predecir).
- **X** es la variable independiente (la característica que se utiliza para hacer la predicción).
- **β₀, β₁, β₂, ..., βₙ** son los coeficientes del polinomio que el modelo debe aprender durante el entrenamiento.
- **ε** es el término de error, que representa la diferencia entre los valores observados y los valores predichos por el modelo.

En la regresión polinomial, el grado del polinomio (n) es un hiperparámetro que se debe ajustar. Un valor más alto de n puede permitir al modelo capturar relaciones no lineales más complejas en los datos, pero también puede llevar a un mayor riesgo de sobreajuste (overfitting) si no se selecciona adecuadamente[[4]](#r4).



**Regresión Polinomial: Predicción polinomial de una función con multiples senos**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# Descripción: En el presente código se creará una función aleatoria conformada por funciones 
# Sinusoidales de diferentes características, se aplica una regresión polinomial con diferentes
# Grados , esto con el objetivo de que el usuario note gráficamente la implicación de la modificación 
# En el grado máximo de la predicción del modelo.

# Se genera los datos de entrada x e y
x = np.arange(0, 10.1, 0.2)
# Se crea una función aleatoria de la cual se obtendrán los datos para entrenar el modelo
y = 2 * np.sin(2 * np.pi * x / 20) + np.sin(2 * np.pi * x / 25) + np.sin(2 * np.pi * x / 5)

# Se convierte x a un arreglo de columna para usarlo en scikit-learn
x = x.reshape(-1, 1)


#########################   G R Á F I C A   1 ############################
# Ajustar el modelo de regresión polinomial
# Se elige el grado máximo del polinomio
grado_pol_1 = 5  
# PolynomialFeatures() genera una  matriz de características que consista en todas las 
# Combinaciones de polinomios de las características con un grado 
# Menor o igual que el grado especificado.
polynomial_features_1 = PolynomialFeatures(degree=grado_pol_1)
# fit_transform(x) realiza simultáneamente operaciones de ajuste y 
# Transformación en los datos de entrada y convierte los puntos de datos.
x_poly_1 = polynomial_features_1.fit_transform(x)
# Se elige el modelo de regresión lineal
model_1 = LinearRegression()
# Se ajustan los parámetros 
model_1.fit(x_poly_1, y)

# Hacer predicciones utilizando el modelo polinomial
y_pred_1 = model_1.predict(x_poly_1)

# Calcular el coeficiente de determinación (R2) y el error cuadrático medio (MSE)
r2_1 = r2_score(y, y_pred_1)
mse_1 = mean_squared_error(y, y_pred_1)

# Graficar los datos y la función polinomial ajustada
plt.scatter(x, y, label='Datos', color='blue')
plt.plot(x, y_pred_1, label='Regresión Polinomial', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Regresión Polinomial de Grado {grado_pol_1}')
plt.legend()
plt.grid(True)
plt.show()

print("Coeficiente de determinación R2:", r2_1)
print("Error cuadrático medio:", mse_1)


#########################   G R Á F I C A   2 ############################
# Ajustar el modelo de regresión polinomial
grado_pol_2 = 15 
polynomial_features_2 = PolynomialFeatures(degree=grado_pol_2)
x_poly_2 = polynomial_features_2.fit_transform(x)

model_2 = LinearRegression()
model_2.fit(x_poly_2, y)

# Hacer predicciones utilizando el modelo polinomial
y_pred_2 = model_2.predict(x_poly_2)

# Calcular el coeficiente de determinación (R2) y el error cuadrático medio (MSE)
r2_2 = r2_score(y, y_pred_2)
mse_2 = mean_squared_error(y, y_pred_2)

# Graficar los datos y la función polinomial ajustada
plt.scatter(x, y, label='Datos', color='blue')
plt.plot(x, y_pred_2, label='Regresión Polinomial', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Regresión Polinomial de Grado {grado_pol_2}')
plt.legend()
plt.grid(True)
plt.show()

print("Coeficiente de determinación R2:", r2_2)
print("Error cuadrático medio:", mse_2)


#########################   G R Á F I C A   3 ############################
# Ajustar el modelo de regresión polinomial
grado_pol_3 = 35 
polynomial_features_3 = PolynomialFeatures(degree=grado_pol_3)
x_poly_3 = polynomial_features_3.fit_transform(x)

model_3 = LinearRegression()
model_3.fit(x_poly_3, y)

# Hacer predicciones utilizando el modelo polinomial
y_pred_3 = model_3.predict(x_poly_3)

# Calcular el coeficiente de determinación (R2) y el error cuadrático medio (MSE)
r2_3 = r2_score(y, y_pred_3)
mse_3 = mean_squared_error(y, y_pred_3)

# Graficar los datos y la función polinomial ajustada
plt.scatter(x, y, label='Datos', color='blue')
plt.plot(x, y_pred_3, label='Regresión Polinomial', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Regresión Polinomial de Grado {grado_pol_3}')
plt.legend()
plt.grid(True)
plt.show()

print("Coeficiente de determinación R2:", r2_3)
print("Error cuadrático medio:", mse_3)
````

De las tres gráficas anteriores es importante notar que el grado máximo del polinomio debe ser ajustado con precaución ya que un grado alto puede llevar a un modelo sobreajustado que no generaliza bien a nuevos datos y causa errores grandes, ajustar el grado a un valor bajo no permite una predicción de los datos satisfactoria y también conlleva su error.Se pueden utilizar técnicas de validación cruzada y curvas de validación para seleccionar el grado óptimo del polinomio y evaluar el rendimiento del modelo.

 dimensiones se mantiene.

En el caso de regresión, el objetivo de SVM es encontrar una función de regresión que se ajuste a los datos con el menor error posible. **SVM: Predecir el valor medio de las viviendas en diferentes áreas de California. Modificado de [California-Housing](https://inria.github.io/scikit-learn-mooc/python_scripts/datasets_california_housing.html)**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Descripción: El siguiente código consiste en un set de datos de sklearn, en este dataset
# Hay precios de casas en california, con múltiples variables independientes y el precio final de la casa
# El objetivo es hacer una predicción de el precio de una casa tomando en cuenta múltiples factores usando SVM.

# Cargar el conjunto de datos California Housing
california_housing = fetch_california_housing(as_frame=True)

# Para obtener una descripción del dataset y sus variables independientes
# Se puede usar: print(california_housing.DESCR), hay 8 variables en total,
# Son:
#         - MedInc        median income in block group
#         - HouseAge      median house age in block group
#         - AveRooms      average number of rooms per household
#         - AveBedrms     average number of bedrooms per household
#         - Population    block group population
#         - AveOccup      average number of household members
#         - Latitude      block group latitude
#         - Longitude     block group longitude

# Se imprimen los primeros datos del frame
print(california_housing.frame.head())
# Datos de características (atributos independientes)
X = california_housing.data  
# Datos de objetivo (atributos dependientes)
y = california_housing.target  
# StandardScaler es una técnica de preprocesamiento que estandariza (normaliza) 
# las características de un conjunto de datos de tal manera que tengan una media de 0 
# y una desviación estándar de 1. Esto significa que después de aplicar la normalización, 
# las características tendrán una escala común
scaler = StandardScaler()
# ajusta el escalador (scaler) a los datos, calculando la media y la desviación estándar 
# de cada característica en el conjunto de entrenamiento.
# Luego, aplica la transformación, restando la media y dividiendo por la desviación estándar 
# para cada característica en X, normalizándolas.
X = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Para obtener los parámetros modificables en svm se puede usar svc.get_params()

# Crear el modelo SVM para regresión, el hiperparámetro C normalmente va de 0.1<C<100
svm_model = SVR(kernel='linear',C=1)

# Entrenar el modelo
svm_model.fit(X_train, y_train)

# Hacer predicciones en los datos de prueba
y_pred = svm_model.predict(X_test)


# Calcular el coeficiente de determinación (R2)
r2 = r2_score(y_test, y_pred)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular las bandas de error
error_std = np.std(y_test - y_pred)
y_upper = y_pred + 1.96 * error_std
y_lower = y_pred - 1.96 * error_std

# Graficar los resultados
plt.scatter(y_test, y_pred)
plt.plot([-1, 5], [-1, 5], 'k--', lw=2)
plt.fill_between(y_pred, y_lower, y_upper, color='red', alpha=0.2, label='Banda de error')
plt.xlabel('Valores reales del precio de las casas')
plt.ylabel('Predicciones del precio de las casas')
plt.title('Regresión SVM - California Housing')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.legend()
plt.show()

print("Coeficiente de determinación R2:", r2)
print("Error cuadrático medio:", mse)
````

La banda de error sombreada podría considerarse como una representación gráfica de la dispersión de las predicciones del modelo. Los puntos de datos reales que están dentro de esta banda podrían considerarse como aquellos para los cuales las predicciones del modelo están dentro del rango de incertidumbre. Por lo tanto, esta banda de error sombreada da una idea de cuán bien se ajustan las predicciones del modelo a los datos reales y cuánta variabilidad hay en las predicciones alrededor de la línea de predicción.

### Construcción del Árbol de Decisión <a class="tocSkip">

Un árbol de decisión se construye de manera recursiva dividiendo el conjunto de datos en subconjuntos más pequeños en función de las características. El objetivo es dividir el conjunto de datos de manera que la información contenida en los subconjuntos sea lo más homogénea posible en términos de la variable objetivo[[12]](#r12) (etiquetas de clasificación o valores de regresión).Para saber más sobre la teória de los árboles de decisión se recomienda la siguiente referencia. [[13]](#r13)
    
    
## Entrenamiento del árbol <a class="tocSkip">

El proceso de entrenamiento de un árbol de predicción (regresión o clasificación) se divide en dos etapas:

1. División sucesiva del espacio de los predictores generando regiones no solapantes (nodos terminales) R1, R2, R3, ..., Rj. Aunque, desde el punto de vista teórico, las regiones podrían tener cualquier forma, si se limitan a regiones rectangulares (de múltiples dimensiones), se simplifica en gran medida el proceso de construcción y se facilita la interpretación.

2. Predicción de la variable respuesta en cada región.

A pesar de la sencillez con la que se puede resumir el proceso de construcción de un árbol, es necesario establecer una metodología que permita crear las regiones R1, R2, R3, ..., Rj, o lo que es equivalente, decidir dónde se introducen las divisiones: en qué predictores y en qué valores de los mismos. Es en este punto donde se diferencian los algoritmos de árboles de regresión y clasificación.

En los árboles de regresión,[[12]](#r12) el criterio empleado con más frecuencia para identificar las divisiones es el Residual Sum of Squares (RSS). El objetivo es encontrar las J regiones (R1, ..., Rj) que minimizan el Residual Sum of Squares (RSS) total:

\[$RSS =  \sum_{j=1}^J \sum_{i∈R_{j}} (yi - ŷR_j)^2 $\]


Donde ŷRj es la media de la variable respuesta en la región Rj. Una descripción menos técnica equivale a decir que se busca una distribución de regiones tal que el sumatorio de las desviaciones al cuadrado entre las observaciones y la media de la región a la que pertenecen sea lo menor posible.



    
### Predicción <a class="tocSkip">

Una vez construido el árbol, se utiliza para realizar predicciones. Dado un nuevo ejemplo, se desciende por el árbol siguiendo las reglas de decisión en cada nodo hasta llegar a una hoja. La etiqueta de clasificación o el valor de regresión en la hoja se utiliza como la predicción.[[12]](#r12)


    
### Beneficios y Consideraciones <a class="tocSkip">

Los árboles de decisión para regresión son flexibles, fáciles de entender e interpretar, y pueden capturar relaciones no lineales en los datos. Sin embargo, también pueden ser propensos al sobreajuste si no se controlan adecuadamente, lo que puede llevar a un rendimiento deficiente en datos nuevos.


**Arboles de Decisión Regresión: predicción de datos lineales**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Generar datos de ejemplo
X, y = make_regression(n_samples=100, n_features=1, noise=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

###########################      G R Á F I C A    1       ####################
# Crear el modelo de árbol de decisión para regresión, se selecciona el hiperparámetro max_depth en 3
regression_tree_1 = DecisionTreeRegressor(max_depth=3)

# Entrenar el modelo
regression_tree_1.fit(X_train, y_train)

# Hacer predicciones en los datos de prueba
y_pred_1 = regression_tree_1.predict(X_test)

# Calcular el coeficiente de determinación (R2)
r2_1 = r2_score(y_test, y_pred_1)

# Graficar el árbol de decisión
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 6))
plot_tree(regression_tree_1, filled=True, feature_names=['Feature'])
plt.title('Árbol de Decisión para Regresión con max_depth=3')
plt.savefig('arbol_regress_1.png')  # Guardar la figura del árbol en un archivo PNG
plt.show()

# Graficar los resultados de la regresión
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Valores reales')
plt.scatter(X_test, y_pred_1, color='red', label='Predicciones')
plt.xlabel('Característica')
plt.ylabel('Valor objetivo')
plt.title('Predicciones de Regresión con Árbol de Decisión con max_depth=3')
plt.legend()
plt.grid(True)
plt.savefig('predicciones_regress_1.png')  # Guardar la figura de las predicciones en un archivo PNG
plt.show()

print("Coeficiente de determinación R2:", r2_1)

###########################      G R Á F I C A    2       ####################
# Crear el modelo de árbol de decisión para regresión
regression_tree = DecisionTreeRegressor(max_depth=7)

# Entrenar el modelo
regression_tree.fit(X_train, y_train)

# Hacer predicciones en los datos de prueba
y_pred = regression_tree.predict(X_test)

# Calcular el coeficiente de determinación (R2)
r2 = r2_score(y_test, y_pred)

# Graficar el árbol de decisión
plt.figure(figsize=(10, 6))
plot_tree(regression_tree, filled=True, feature_names=['Feature'])
plt.title('Árbol de Decisión para Regresión con max_depth=7')
plt.savefig('arbol_regress_2.png')  # Guardar la figura del árbol en un archivo PNG
plt.show()

# Graficar los resultados de la regresión
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Valores reales')
plt.scatter(X_test, y_pred, color='red', label='Predicciones')
plt.xlabel('Característica')
plt.ylabel('Valor objetivo')
plt.title('Predicciones de Regresión con Árbol de Decisión con max_depth=7')
plt.legend()
plt.grid(True)
plt.savefig('predicciones_regress_2.png')  # Guardar la figura de las predicciones en un archivo PNG
plt.show()

print("Coeficiente de determinación R2:", r2)
````

Como se puede notar, al ajustar el valor de profundidad máximo se puede mejorar considerablemente la precisión de las predicciones realizadas sin embargo la representación gráfica del árbol se vuelve poco manejable, la elección del max_depth debe tener en cuenta tanto la precisión y la capacidad de entendimiento del árbol como tal, así como evitar el sobreajuste del modelo.

## **Entropía** <a class="tocSkip">
    
    
La entropía es una medida de la impureza o la falta de homogeneidad en un conjunto de datos. En el contexto de los árboles de decisión de clasificación, se utiliza para cuantificar cuán mezcladas están las clases en un nodo. La fórmula de entropía se define como[[15]](#r15):
    
    
    
    Entropía(S) = -Σ(pi * log2(pi))

    
    
Donde:
- Entropía(S) es la entropía del conjunto S.
    
- c es el número de clases en el conjunto S.
    
- pi es la proporción de ejemplos en la clase i en el conjunto S.
    
    
Cuando la entropía es baja (cerca de 0), significa que el nodo es más puro, es decir, contiene principalmente ejemplos de una sola clase. Cuando la entropía es alta (cerca de 1), significa que el nodo es impuro y contiene una mezcla de clases.
    
    
## **Impureza de Gini** <a class="tocSkip">
    
    
La impureza de Gini es la probabilidad de clasificar incorrectamente un punto de datos aleatorio en el conjunto de datos si se etiquetara en función de la distribución de clases del conjunto de datos[[15]](#r15). La fórmula de la impureza de Gini se define como:
    
    
    Impureza de Gini(S) = 1 - Σ(pi²)

Donde:
    
- Impureza de Gini(S) es la impureza de Gini del conjunto S.
    
- c es el número de clases en el conjunto S.
    
- pi es la proporción de ejemplos en la clase i en el conjunto S.
    
    
La impureza de Gini también mide cuán mezcladas están las clases en un nodo. Al igual que con la entropía, cuanto más cercano a 0 sea el valor de Gini, más puro será el nodo, y cuanto más cercano a 0.5 (el valor máximo), más impuro será el nodo.


###  Random Forest

Random Forest es un algoritmo de aprendizaje automático que se utiliza tanto para tareas de clasificación como de regresión. Es una técnica de **ensamble** que combina múltiples árboles de decisión individuales para crear un modelo más robusto y preciso. Cada árbol en el conjunto se construye a partir de una muestra aleatoria del conjunto de datos de entrenamiento y produce una predicción[[18]](#r18). Luego, las predicciones individuales de los árboles se combinan para obtener una predicción final más confiable.

### Principales Características: <a class="tocSkip">

1. **Ensamble de Árboles de Decisión:** Random Forest se compone de múltiples árboles de decisión, llamados "árboles de decisión base". Cada árbol se entrena en una muestra aleatoria del conjunto de datos de entrenamiento, lo que introduce variabilidad y diversidad en los modelos individuales.

2. **Muestreo Aleatorio:** Para crear diferentes conjuntos de entrenamiento para cada árbol, se utiliza un proceso llamado "muestreo con reemplazo" (bootstrap) en el que se selecciona una muestra aleatoria del conjunto de datos original. Algunas observaciones pueden repetirse en la muestra, mientras que otras pueden quedar fuera.

3. **Votación o Promedio:** Una vez que se entrenan todos los árboles, las predicciones individuales de cada árbol se combinan para producir una predicción final. En la clasificación, se puede usar la "votación mayoritaria", donde la clase más frecuente se elige como la predicción final. En la regresión, se promedian las predicciones de los árboles.

4. **Reducción de la Correlación:** Random Forest reduce la correlación entre los árboles al introducir aleatoriedad en la construcción de cada árbol. Esto ayuda a prevenir el sobreajuste y mejora la generalización del modelo.

5. **Importancia de Características:** Random Forest proporciona una medida de la importancia de las características al observar cuánto contribuye cada característica a la mejora de la precisión del modelo. Esta importancia puede ser útil para la selección de características.
    
## Hiperparámetros Clave <a class="tocSkip">

Los tres hiperparámetros[[18]](#r18) clave en Random Forest son:

1. Número de árboles (n_estimators): Controla cuántos árboles se construyen en el bosque.
2. Profundidad máxima del árbol (max_depth): Limita la profundidad máxima de cada árbol para evitar el sobreajuste.
3. Número de características a considerar en cada división (max_features): Define cuántas características se consideran al realizar divisiones en cada nodo del árbol.


### Ventajas y Limitaciones: <a class="tocSkip">

**Ventajas:**
- Puede manejar problemas de clasificación y regresión.
- Robusto frente a valores atípicos y datos ruidosos.
- Menos propenso al sobreajuste en comparación con árboles individuales.
- Capacidad para manejar una gran cantidad de características.
- Proporciona una estimación de la importancia de las características.

**Limitaciones:**
- Puede requerir más tiempo de entrenamiento y recursos computacionales debido a la construcción de múltiples árboles.
- La interpretación del modelo puede ser más complicada debido a la combinación de varios árboles.

### Regresión Logística

La regresión logística es un algoritmo de aprendizaje supervisado que se utiliza principalmente para la clasificación de datos binarios. El objetivo de la regresión logística es encontrar una relación matemática entre un conjunto de variables de entrada (características o atributos) y la probabilidad de que un evento ocurra o pertenezca a una de las dos clases[[19]](#r19). La relación matemática se establece a través de la función logística, que toma una combinación lineal de las variables de entrada y la transforma en un valor entre 0 y 1. Esta transformación es importante porque se utiliza para estimar la probabilidad de que un punto de datos pertenezca a una de las clases.


¿Qué ocurre cuando la variable respuesta es binaria, solo dos posibles valores?

Si una variable cualitativa con dos niveles (binaría) se codifica como 0 y 1, matemáticamente es posible ajustar un modelo de regresión lineal por mínimos cuadrados. Sin embargo, esta aproximación tiene dos problemas:

-  generar una recta (hiperplano si hay múltiples variables), se pueden obtener valores predichos distintos de 0 y 1, lo que entra en contradicción con la definición de la variable respuesta binaria.

-  se quiere interpretar las predicciones del modelo como probabilidades de pertenencia a cada clase, no se cumpliría la condición de que toda probabilidad ha de estar dentro del intervalo [0,1], ya que podrían obtenerse valores fuera de este.

### Principales Características: <a class="tocSkip">

1. **Función Logística (Sigmoidal):** La Regresión Logística utiliza la función logística (o función sigmoide) para modelar la relación entre las características y la probabilidad de pertenecer a una clase[[19]](#r19). La función logística tiene la forma:
    
    
    $sigmoide=\sigma(x) = \frac{1}{1 + e^{-x}}$
    
donde \(e\) es la base del logaritmo natural.

2. **Probabilidad Estimada:** El resultado de la Regresión Logística es la probabilidad estimada de que una instancia pertenezca a la clase positiva (1). Para realizar la clasificación, se establece un umbral de probabilidad, por encima del cual se clasifica como la clase positiva y por debajo como la clase negativa[[19]](#r19).

### Ventajas y Limitaciones: <a class="tocSkip">
**Ventajas:**
- Adecuada para problemas de clasificación binaria y multiclase.
- Proporciona probabilidades como salida, lo que es útil para tomar decisiones ponderadas.
- Es menos propensa al sobreajuste en comparación con modelos más complejos.
- Puede manejar características numéricas y categóricas.

**Limitaciones:**
- No es adecuada para problemas de regresión (predicción de valores continuos).
- Puede tener dificultades con conjuntos de datos con características altamente correlacionadas.

### Fórmula Matemática: <a class="tocSkip">

La Regresión Logística modela la probabilidad \(P(y=1 $\mid x$\) de que una instancia \(x\) pertenezca a la clase positiva (1)[[19]](#r19). La relación entre las características \(x\) y la probabilidad logarítmica se modela mediante una combinación lineal:

\[$ \log \left( \frac{P(y=1 \mid x)}{1 - P(y=1 \mid x)} \right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n$ \]

donde \( $\beta_0, \beta_1, \ldots, \beta_n $\) son los coeficientes de la regresión.

### Vecinos Cercanos (KNN) para Clasificación
El algoritmo de k vecinos más cercanos, también conocido como KNN o k-NN, es un clasificador de aprendizaje supervisado no paramétrico, que utiliza la proximidad para hacer clasificaciones o predicciones sobre la agrupación de un punto de datos individual.[[21]](#r21)
![arbol decisión](https://helloacm.com/wp-content/uploads/2016/03/2012-10-26-knn-concept.png)
<span style="display: block; text-align: center;">Tomado de [[20]](#r20)</span>  


La imagen anterior  muestra un diagrama de dispersión (nube de puntos). En el diagrama existen dos tipos de clases (Clase A y Clase B). Lo que se busca, es que al ingresar un nuevo punto, este se clasifica en una de las clases (ya sea en la clase A o en la clase B). Cuando se ingresa un nuevo dato (la estrella central), K-NN empleará su procedimiento de clasificación para asignar a este dato nuevo, una clase. ¿Mediante que procedimiento?[[20]](#r20) Buscar a los vecinos más cercanos y de este modo clasificar a ese dato (estrella). En la figura se observa una variable llamada K, esta variable contiene el número de vecinos que el clasificador debe buscar en el diagrama. Se observa que en el momento en el que k=3, se crea un tipo de circunferencia al rededor de la estrella. Dicha circunferencia encierra a los 3 puntos más próximos a la estrella. Entre estos 3 puntos, 2 de ellos son de la clase B y una pertenece a la clase A. K-nn, clasificará la estrella según el número de vecinos más cercanos (en este caso, a la clase B). Pero si se cambia el valor de la variable K a 6, la predicción cambia ya que entre sus 6 vecinos más cercanos, 4 de ellos pertenecen a la clase A, y solo 2 pertenecen a al clase B.
### Principales Características <a class="tocSkip">
1. **Medida de Distancia**
KNN utiliza una medida de distancia, como la distancia euclidiana[[21]](#r21), para calcular la proximidad entre las instancias. Las instancias más cercanas se consideran los "vecinos" y se utilizan para tomar decisiones de clasificación.

2. **Elección del Parámetro K**
El parámetro 
K en KNN determina cuántos vecinos se consideran al hacer una predicción. Si 
K es pequeño, el modelo puede ser más sensible al ruido en los datos. Si 
K es grande, la decisión puede verse afectada por la mayoría de los puntos cercanos.

3. **Clasificación por Mayoría**
En la clasificación KNN, una instancia se asigna a la clase que sea la mayoría entre sus 
K vecinos más cercanos. La clase se determina por votación.

### **Ventajas y Limitaciones** <a class="tocSkip">

**Ventajas:**
- Simple de entender e implementar.
- Puede manejar problemas de clasificación con múltiples clases.
- Es capaz de capturar relaciones complejas en los datos.


**Limitaciones:**
- Sensible a la elección de K y la medida de distancia.
- Puede ser computacionalmente costoso en conjuntos de datos grandes.
- No es eficiente para conjuntos de datos con muchas características.

# Selección de características


La selección de características es un proceso crítico en el desarrollo de modelos de aprendizaje automático. Se refiere a identificar y elegir las características más relevantes e informativas del conjunto de datos que se utilizarán para entrenar el modelo. La importancia de la selección de características se puede destacar por las siguientes razones:

**Mejora del rendimiento del modelo:** Al seleccionar solo las características más relevantes, se pueden eliminar características redundantes o ruidosas que no aportan información útil al modelo. Esto puede conducir a una mejora en el rendimiento del modelo, ya que se puede enfocar en patrones más significativos.

**Reducción de la complejidad del modelo:** Al reducir el número de características, se puede disminuir la complejidad del modelo, lo que puede ayudar a evitar el sobreajuste (overfitting) y a mejorar la generalización del modelo a nuevos datos.

**Reducción del tiempo de entrenamiento:** Un conjunto de datos con muchas características puede requerir más tiempo de entrenamiento. Al seleccionar un subconjunto relevante de características, se puede reducir el tiempo de entrenamiento sin perder información crucial.

**Interpretación del modelo:** Al utilizar un subconjunto más pequeño y relevante de características, el modelo resultante puede ser más fácil de interpretar y explicar. Esto es especialmente importante en aplicaciones donde se requiere una comprensión clara de cómo las características afectan las predicciones del modelo.

**Manejo de características faltantes o ruido:** Al seleccionar solo características relevantes, se puede mejorar la capacidad del modelo para manejar datos faltantes o ruido en el conjunto de datos.


En general hay dos tipos de selección de datos: métodos de filtro y métodos de envoltura

### Métodos de filtro:

Los métodos de filtro son métodos de selección de características que se aplican antes de entrenar el modelo. Estos métodos se basan en medidas estadísticas o heurísticas para evaluar la relevancia de las características en relación con la variable objetivo. Algunos ejemplos de métodos de filtro incluyen:

1. **Correlación:** Se calcula la correlación entre cada característica y la variable objetivo. Las características con una alta correlación se consideran más relevantes.

2. **Pruebas estadísticas:** Se utilizan pruebas estadísticas como ANOVA o t-prueba para evaluar si hay diferencias significativas en las características entre diferentes clases de la variable objetivo.

3. **Mutual Information:** Calcula la información mutua entre cada característica y la variable objetivo para medir la dependencia entre ellos.

4. **Varianza:** Se seleccionan características con una varianza por encima de un umbral determinado. Las características con poca variabilidad pueden no aportar información útil.

### Métodos de envoltura:

Los métodos de envoltura son métodos de selección de características que se aplican mientras se entrena el modelo. Estos métodos utilizan un algoritmo de aprendizaje automático para evaluar la importancia de las características y seleccionar un subconjunto óptimo. Algunos ejemplos de métodos de envoltura incluyen:

1. **Forward Selection:** Comienza con un conjunto vacío de características y agrega una característica a la vez, seleccionando la que mejor mejora el rendimiento del modelo.

2. **Backward Elimination:** Comienza con todas las características y elimina una característica a la vez, seleccionando la que menos afecta negativamente el rendimiento del modelo.

3. **Recursive Feature Elimination (RFE):** Es un método iterativo que selecciona características en función de los coeficientes de importancia proporcionados por un modelo de aprendizaje automático.

4. **Algoritmos genéticos:** Utilizan estrategias de búsqueda inspiradas en la evolución para encontrar el subconjunto óptimo de características.

Es importante tener en cuenta que tanto los métodos de filtro como los métodos de envoltura tienen sus ventajas y desventajas. Los métodos de filtro son más rápidos y menos costosos computacionalmente, pero pueden pasar por alto interacciones complejas entre características. Por otro lado, los métodos de envoltura pueden ser más precisos pero son computacionalmente más costosos. La elección del método de selección de características dependerá del problema específico y del conjunto de datos en cuestión.

## Hiperparámetro: <a class="tocSkip"> 
Los hiperparámetros son variables de configuración externa que los científicos de datos utilizan para administrar el entrenamiento de modelos de machine learning. A veces llamados hiperparámetros de modelos, los hiperparámetros se configuran de manera manual antes de entrenar un modelo[[9]](#r9). Los parámetros no están sujetos a programación por parte de científicos de datos, sino que se derivan de componentes internos durante el proceso de aprendizaje. El uso de hiperparámetros identifica factores cruciales como la arquitectura del modelo, la tasa de aprendizaje y la complejidad del modelo.Algunos ejemplos de hiperparámetros son:
    
    
- Árboles de decisión (max_depth): Controla la profundidad máxima del árbol.
- Árboles de decisión (min_samples_leaf): Establece el número mínimo de muestras requeridas en un nodo hoja.
- Redes Neuronales (hidden_layer_sizes):Número de capas y unidades ocultas  Define la arquitectura de la red.
- Redes Neuronales (learning_rate): Controla la velocidad de aprendizaje.
- K-Vecinos más Cercanos (n_neighbors): Define el número de vecinos considerados.
    
 <a name="t2"></a>
## Función de pérdida: <a class="tocSkip">  
En el contexto del aprendizaje automático (machine learning), una función de pérdida  es una métrica que se utiliza para evaluar cuán bien se está desempeñando un modelo en la tarea que está diseñado para llevar a cabo. Su función principal es cuantificar la discrepancia entre las predicciones del modelo y los valores reales de los datos[[16]](#r16).

La función de pérdida juega un papel crucial en el proceso de entrenamiento de un modelo de machine learning. Durante el entrenamiento, el objetivo es ajustar los parámetros del modelo de manera que la función de pérdida se minimice. En otras palabras, se busca encontrar los parámetros del modelo que produzcan predicciones que estén lo más cerca posible de los valores reales de los datos.
    

 <a name="t3"></a>
## Matriz de confusión: <a class="tocSkip">  
Es una herramienta utilizada para evaluar el rendimiento de un modelo de clasificación en machine learning. Proporciona una visión detallada de cómo el modelo clasifica las instancias de un conjunto de datos en función de las clases reales a las que pertenecen[[17]](#r17). La matriz de confusión se compone de cuatro componentes principales:

Verdaderos Positivos (True Positives, TP): Representa el número de instancias que fueron correctamente clasificadas como positivas por el modelo. En otras palabras, son los casos en los que el modelo predijo correctamente una instancia como perteneciente a la clase positiva, y la instancia realmente pertenece a esa clase.

Verdaderos Negativos (True Negatives, TN): Representa el número de instancias que fueron correctamente clasificadas como negativas por el modelo. Estos son los casos en los que el modelo predijo correctamente que una instancia no pertenece a la clase positiva, y la instancia realmente no pertenece a esa clase.

Falsos Positivos (False Positives, FP): Representa el número de instancias que fueron incorrectamente clasificadas como positivas por el modelo, cuando en realidad pertenecen a la clase negativa. Estos son errores de clasificación en los que el modelo "falsamente" predijo la clase positiva.

Falsos Negativos (False Negatives, FN): Representa el número de instancias que fueron incorrectamente clasificadas como negativas por el modelo, cuando en realidad pertenecen a la clase positiva. Estos son errores de clasificación en los que el modelo "falsamente" predijo la clase negativa.

La matriz de confusión se presenta generalmente en una tabla con dos filas y dos columnas (para problemas de clasificación binaria), y se ve de la siguiente manera:

Estas métricas ayudan a comprender cómo el modelo está clasificando las instancias y a evaluar su capacidad para predecir correctamente las clases de interés.
